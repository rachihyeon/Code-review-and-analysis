```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline 
!pip install koreanize-matplotlib
import koreanize_matplotlib
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
plt.rc('font',family='Malgun Gothic')
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
def print_plot(dataframe):
    x = dataframe['N']
    y = dataframe['time']
    fig = plt.figure()
    fig.set_facecolor('white')
    sns.lineplot(x=x, y=y, color='r', marker='o')
    
def merge_plot(dataframe_array):
    labels = ['naive', 'Backtracking', 'Bitwise', 'Z3']
    fig = plt.figure()
    fig.set_facecolor('white')
    for dataframe, label in zip(dataframe_array, labels):
        x = dataframe['N']
        y = dataframe['time']
        sns.lineplot(x=x, y=y, label=label)
        plt.legend()
```
